Importa los datos generando los grafos para cada capa, y obtén el número de nodos y aristas para cada una de ellas. **(0,25 puntos)**

```python
# Ruta donde están los CSVs
data_path = "data"

# Inicializar diccionario para almacenar los grafos de cada temporada
season_graphs = {}

# Iterar por las temporadas 1 a 8
for season in range(1, 9):
    nodes_file = os.path.join(data_path, f"got-s{season}-nodes.csv")
    edges_file = os.path.join(data_path, f"got-s{season}-edges.csv")
    
    if os.path.exists(nodes_file) and os.path.exists(edges_file):
        # Cargar archivos CSV
        nodes_df = pd.read_csv(nodes_file)
        edges_df = pd.read_csv(edges_file)
        
        # Crear grafo vacío para la temporada
        G = nx.Graph()
        G.add_nodes_from(nodes_df['Id'])
        
        # Añadir aristas y nodos si faltan según el enunciado
        for _, row in edges_df.iterrows():
            source = row['Source']
            target = row['Target']
            weight = row.get('Weight', 1)
            
            if source not in G:
                G.add_node(source)
            if target not in G:
                G.add_node(target)
            G.add_edge(source, target, weight=weight)
        
        # Guardar el grafo en el diccionario
        season_graphs[season] = G

# Mostrar resumen
for season, G in season_graphs.items():
    print(f"Temporada {season}: {G.number_of_nodes()} nodos, {G.number_of_edges()} aristas")
```

```python
# Parámetros de visualización
layer_spacing = 10  # Espacio horizontal entre temporadas
node_color = "#1f78b4"
edge_color = "gray"
node_size = 50
alpha_edges = 0.3
alpha_nodes = 0.9

# Crear figura
plt.figure(figsize=(20, 8))

# Diccionario para almacenar posiciones por nodo por temporada
node_positions = {}

# Visualizar cada grafo con desplazamiento horizontal por temporada
for season, G in season_graphs.items():
    # Calcular layout
    pos = nx.spring_layout(G, seed=season)  # Fijar semilla para reproducibilidad

    # Desplazar nodos horizontalmente según la temporada
    for node in pos:
        pos[node][0] += (season - 1) * layer_spacing

    # Guardar posición para posibles enlaces intercapa
    node_positions[season] = pos

    # Dibujar nodos y aristas
    nx.draw_networkx_edges(G, pos, edge_color=edge_color, alpha=alpha_edges)
    nx.draw_networkx_nodes(G, pos, node_color=node_color, alpha=alpha_nodes, node_size=node_size)

# Dibujar enlaces intercapa entre mismos nodos en temporadas consecutivas
for season in range(1, 8):  # De la temporada 1 a la 7
    G1 = season_graphs[season]
    G2 = season_graphs[season + 1]
    common_nodes = set(G1.nodes).intersection(G2.nodes)
    
    for node in common_nodes:
        pos1 = node_positions[season][node]
        pos2 = node_positions[season + 1][node]
        plt.plot([pos1[0], pos2[0]], [pos1[1], pos2[1]], color="black", linewidth=0.5, alpha=0.1)

# Ajustes finales
plt.title("Red multilayer de Juego de Tronos (8 temporadas)")
plt.axis('off')
plt.tight_layout()

# Guardar como imagen
plt.savefig("grafo_multilayer_got.png", dpi=300, bbox_inches="tight")
plt.show()
```

La entropía de Von Neumann es una medida inspirada en la teoría cuántica que se puede usar para cuantificar cuán compleja o estructurada es una red. Para calcularla en un grafo, seguimos estos pasos:

Normalizamos la matriz de Laplaciana del grafo:
Esto nos da una representación de cómo están conectados los nodos entre sí, ajustando por el grado de cada nodo. Es decir, no solo importa quién está conectado a quién, sino cuántas conexiones tiene cada nodo en total.

Estos valores representan propiedades fundamentales de la red, como su conectividad global.

Ignoramos los autovalores que sean cero para evitar errores de logaritmo.
El resultado es una sola cifra que resume el grado de “desorden” o “diversidad estructural” del grafo.

✅ ¿Qué nos indica esta entropía?

Una entropía baja suele indicar una red más ordenada o redundante (más predecible).

Una entropía alta indica mayor complejidad, diversidad estructural o más incertidumbre en la organización de la red.

```python
# Calcular la entropía de Von Neumann por temporada

def von_neumann_entropy(G):
    # Matriz Laplaciana normalizada
    L = nx.normalized_laplacian_matrix(G).todense()
    # Autovalores
    eigenvalues = np.linalg.eigvalsh(L)
    # Filtrar valores pequeños o negativos por estabilidad numérica
    eigenvalues = np.clip(eigenvalues, 1e-12, 1)
    # Entropía de Von Neumann
    entropy = -np.sum(eigenvalues * np.log(eigenvalues))
    return entropy

# Calcular entropía para cada grafo de temporada
vn_entropies = {}

for season, G in season_graphs.items():
    entropy = von_neumann_entropy(G)
    vn_entropies[season] = entropy
    print(f"Temporada {season}: Entropía de Von Neumann = {entropy:.4f}")
```

```python
# Lista de todas las temporadas
temporadas = list(season_graphs.keys())

# Obtener todos los personajes que aparecen en alguna temporada
todos_los_nodos = set()
for G in season_graphs.values():
    todos_los_nodos.update(G.nodes())

# Crear un DataFrame donde las filas son personajes y columnas temporadas
df_centralidad = pd.DataFrame(index=list(todos_los_nodos), columns=temporadas)


# Rellenar el DataFrame con la centralidad de grado por temporada
for temporada in temporadas:
    G = season_graphs[temporada]
    grados = dict(G.degree())
    for nodo in todos_los_nodos:
        df_centralidad.loc[nodo, temporada] = grados.get(nodo, 0)

# Convertir a float por si hay valores NaN o strings
df_centralidad = df_centralidad.astype(float)

# Calcular la desviación estándar por nodo (a través de temporadas)
df_centralidad['std'] = df_centralidad.std(axis=1)

# Opcional: ordenar por estabilidad (de menor a mayor std)
df_estabilidad = df_centralidad.sort_values('std')

# Mostrar los 10 personajes más estables y más inestables
print("🔒 Personajes más estables (std más baja):")
print(df_estabilidad['std'].head(10))
print("\n⚠️ Personajes más inestables (std más alta):")
print(df_estabilidad['std'].tail(10))

# 📊 Visualización
plt.figure(figsize=(12, 5))
plt.hist(df_centralidad['std'], bins=30, color="#69b3a2", edgecolor="black", alpha=0.8)
plt.xlabel("Desviación estándar del grado (variación estructural)")
plt.ylabel("Número de personajes")
plt.title("Distribución de estabilidad estructural de personajes en Juego de Tronos")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Esto se refleja en el pico del histograma entre 0 y 2, donde se concentran la mayoría de los personajes.

Son personajes secundarios o de fondo que aparecen consistentemente con roles menores o periféricos.

```python
# Construir el grafo multicapa combinado
multi_graph = nx.Graph()
for G in season_graphs.values():
    multi_graph = nx.compose(multi_graph, G)
```

```python

# Simulación de percolación dirigida y aleatoria
def simulate_percolation(G, strategy='random'):
    G_copy = G.copy()
    num_nodes = G_copy.number_of_nodes()
    removed = 0
    giant_component_sizes = []

    if strategy == 'degree':
        nodes_sorted = sorted(G_copy.degree, key=lambda x: x[1], reverse=True)
        nodes_to_remove = [node for node, _ in nodes_sorted]
    elif strategy == 'random':
        nodes_to_remove = list(G_copy.nodes())
        random.shuffle(nodes_to_remove)
    else:
        raise ValueError("Estrategia debe ser 'degree' o 'random'.")

    for node in nodes_to_remove:
        G_copy.remove_node(node)
        removed += 1
        giant_size = len(max(nx.connected_components(G_copy), key=len)) if G_copy.number_of_nodes() > 0 else 0
        giant_component_sizes.append(giant_size / num_nodes)  # % tamaño comp. gigante

    return [i / num_nodes for i in range(num_nodes)], giant_component_sizes
```

```python
# Ejecutar simulaciones
x_deg, y_deg = simulate_percolation(multi_graph, strategy='degree')
x_rand, y_rand = simulate_percolation(multi_graph, strategy='random')

# Visualización
plt.figure(figsize=(10, 6))
plt.plot(x_deg, y_deg, label="Percolación dirigida (grado)", color='crimson')
plt.plot(x_rand, y_rand, label="Percolación aleatoria", color='steelblue')
plt.axvline(x_deg[next(i for i, y in enumerate(y_deg) if y < 0.5)], linestyle='--', color='crimson', alpha=0.6)
plt.axvline(x_rand[next(i for i, y in enumerate(y_rand) if y < 0.5)], linestyle='--', color='steelblue', alpha=0.6)

plt.xlabel("Fracción de nodos eliminados")
plt.ylabel("Fracción de nodos en componente gigante")
plt.title("Percolación intercapa en red de Juego de Tronos")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

En esta sección analizamos cómo se comporta la red multicapa de Juego de Tronos ante procesos de percolación, es decir, eliminación progresiva de nodos. Hemos simulado dos escenarios:

🔴 Percolación dirigida por grado agregado: Se eliminan primero los personajes con mayor conectividad total en la red (los más centrales).

🔵 Percolación aleatoria: Los personajes se eliminan en orden aleatorio, simulando un fallo estructural no dirigido.

La curva roja (percolación dirigida) cae abruptamente: eliminando aproximadamente un 17% de los nodos más conectados, la componente gigante de la red colapsa casi por completo.

La curva azul (percolación aleatoria) desciende más gradualmente: se requiere eliminar más del 40% de los nodos aleatoriamente para alcanzar una fragmentación comparable.

Esto muestra que la red es muy vulnerable a ataques dirigidos a los nodos con mayor grado, pero relativamente robusta ante fallos aleatorios.

## Complejos Simpliciales en contagios sociales (3,5 puntos)

El artículo "Simplicial models of social cotagion" de Lacopini et al. (2019), introduce una metodología para el estudio de procesos de contagios que va más allá de las interacciones binarias, grafos simples, introduciendo interacciones de orden superior mediante complejos simpliciales. 

Este artículo será la base de este segundo y último ejercicio, pidiéndote que recrees la Figura 3 de dicho artículo la cual muestra la dinámica del Modelo implementado (SCM o Simplicial Contagion Model), sobre un Complejo Simplicial Aleatorio (Random Simplicial Complex) o RSC. Vamos a ello.

Basándote en el artículo en cuestión, elabora el código que permita ejecutar el SCM sobre un RSC con los mismos parámetros que los de dicho artículo. Comprueba el modelo obtenido imprimiendo resultados de algún parámetro y/o ejecutando una simulación corta. 

<div style="background-color: #fcf2f2; border-color: #dfb5b4; border-left: 5px solid #dfb5b4; padding: 0.5em;">
<strong>Solución</strong>
</div>

```python
def generate_random_simplicial_complex(N, p, p_tri):
    G = nx.erdos_renyi_graph(N, p)
    triangles = set()
    nodes = list(G.nodes)
    for trio in combinations(nodes, 3):
        u, v, w = trio
        if G.has_edge(u, v) and G.has_edge(v, w) and G.has_edge(u, w):
            if random.random() < p_tri:
                triangles.add(trio)
    return G, triangles

```

```python
def run_SCM(G, triangles, beta, beta_tri, T, mu=1.0, initial_infected_size=10):
    state = {node: 0 for node in G.nodes}
    initial_infected = random.sample(list(G.nodes()), k=initial_infected_size)
    for node in initial_infected:
        state[node] = 1

    prevalence = []

    for _ in range(T):
        new_state = state.copy()
        for node in G.nodes:
            if state[node] == 0:  # Susceptible
                infected_neighbors = [nbr for nbr in G.neighbors(node) if state[nbr] == 1]
                prob_inf = 1 - (1 - beta)**len(infected_neighbors)

                for tri in triangles:
                    if node in tri:
                        others = [n for n in tri if n != node]
                        if all(state[n] == 1 for n in others):
                            prob_inf = 1 - (1 - prob_inf) * (1 - beta_tri)

                if random.random() < prob_inf:
                    new_state[node] = 1

            elif state[node] == 1:  # Infectado
                if random.random() < mu:
                    new_state[node] = 0

        state = new_state
        infected_count = sum(state.values())
        prevalence.append(infected_count / len(G.nodes))

    return prevalence
```

```python
# Parámetros
N = 1000
p = 0.01
p_tri = 0.01
T = 100
beta = 0.06
beta_tri = 0.4

# Generar complejo simplicial y simular
G, triangles = generate_random_simplicial_complex(N, p, p_tri)
prevalence = run_SCM(G, triangles, beta, beta_tri, T)

# Visualizar
plt.figure(figsize=(10, 5))
plt.plot(prevalence, label=f"β={beta}, βΔ={beta_tri}")
plt.xlabel("Tiempo")
plt.ylabel("Prevalencia (fracción infectada)")
plt.title("Simulación SCM sobre Complejo Simplicial Aleatorio")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

✅ Interpretación del resultado:

Eje Y: fracción de nodos infectados.

Eje X: tiempo (número de iteraciones o pasos).

Curva: muestra cómo evoluciona la infección en el tiempo para el caso con parámetros β = 0.06 y β_Δ = 0.4.

Esto indica que el contagio no se sostiene en el sistema con estos parámetros: la enfermedad (o idea, rumor, etc.) no se propaga y desaparece rápidamente.
